```python
#Bivariate Analysis
df.corr()#find co-relation
```

```python
#use hitmap to plot the relation between the available parameters
plt.figure(figsize =(18,12))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
#regression plot for price vs living area(sqft)
sns.regplot(x = 'price',y = 'sqft_living',data = df , x_jitter=0.2, scatter_kws={'alpha':0.2})
sns.despine()
```

```python
#regression plot for number of bedrooms vs price:
sns.regplot(x = 'bedrooms',y='price',data = df , x_jitter=0.3, scatter_kws={'alpha':0.3}, color = 'red')
sns.despine()
```

```python
#regression plot for price vs latitude
sns.regplot(x = 'lat',y = 'price',data = df , x_jitter=0.2, scatter_kws={'alpha':0.2},color='black')
sns.despine()
```

```python
#regression plt for price vs longitude:
sns.regplot(x = 'long',y = 'price',data = df , x_jitter=0.2, scatter_kws={'alpha':0.2},color='purple')
sns.despine()
```

```python
#regression plot for condition vs built year
sns.regplot(x = 'condition',y = 'yr_built',data = df , x_jitter=0.2, scatter_kws={'alpha':0.2},color='grey')
sns.despine()
```

```python
#boxplot for price vs view:
sns.boxplot(x = 'view',y = 'price',data = df )
sns.despine()
```

```python
#boxplot for price vs waterfront:
sns.boxplot(x = 'waterfront',y = 'price',data = df)
sns.despine()
```

```python
#Multivariate Analysis
#facet grid for price vs bedrooms
grid = sns.FacetGrid(df, col='condition',col_wrap=2)
grid.map(plt.scatter,'bedrooms','price',alpha = 0.2)
sns.despine()
```

```python
 #facet grid for price vs living area in terms of condition:
grid = sns.FacetGrid(df, col='condition',col_wrap=2)
grid.map(plt.scatter,'sqft_living','price',alpha = 0.2)
sns.despine()
```

```python
#facet grid for price vs view in terms of condition:
grid = sns.FacetGrid(df, col='condition',col_wrap=2)
grid.map(plt.scatter,'view','price',alpha = 0.2)
sns.despine()
```

```python
sns.boxplot(x = 'view',y = 'price',data = df )
sns.despine()
```

```python
#Living Area vs Price in terms of Condition- Facet Grid
grid = sns.FacetGrid(df, col='condition',col_wrap=2)
grid.map(plt.scatter,'sqft_living','price',alpha = 0.2)
sns.despine()
```

```python
#FINAL OBSERVATION:
Better the condition, more the price.
More the views arond, more the price.
More the number of bedrooms, more the price.
More the number of bathrooms, more the price.
More the living area, more the price.
Later the built year, more the price.
If waterfront, more the price.
If more lot area, more the price.
```
